```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = .5
r = .5
p = 2/3
points = 1000
r_true = []
r_in = []
r_out = []
r_in_sigma = []
r_out_sigma = []
dominio = np.linspace(0,1.5,points)
for g in dominio:
  aux = (g*g)/(p*(1+r))
  r_true.append(aux)
  r_in.append(aux*(1-q))
  r_out.append(aux/(1-q))
  r_in_sigma.append(aux*(1-(r*q)/(r+q*(1+r))))
  r_out_sigma.append(aux*(1+(r*q)/(r+q)))
plt.plot(r_true,dominio,label="R^2_true")
plt.plot(r_in,dominio,label="R^2_in")
plt.plot(r_out,dominio,label="R^2_out")
plt.plot(r_in_sigma,dominio,label="R^2_in_sigma")
plt.plot(r_out_sigma,dominio,label="R^2_out_sigma")
plt.xlim(0,2)
plt.legend(loc='upper left',bbox_to_anchor = (1, .975)) 
```

```python
q = .5
p = 100
n = 200
sigma = .2
g = np.ones(p)
g = np.array(g).reshape(1,p)
H_1 = np.random.randn(p,n)*np.sqrt(sigma)
H_2 = np.random.randn(p,n)*np.sqrt(sigma)
Cov1 = (1/n)*H_1.dot(np.transpose(H_1))
Cov2 = (1/n)*H_2.dot(np.transpose(H_2))
H_Wishart = np.concatenate((H_1,H_2),axis=1)
Hsy_Wishart = (1/n)*H_Wishart.dot(np.transpose(H_Wishart))
```

```python
dominio = np.linspace(0,100,1000)
r_true = []
r_in = []
r_out = []
inv_in = np.linalg.inv(Cov1)
deno = g.dot(np.linalg.inv(Hsy_Wishart).dot(np.transpose(g)))
deno_in = g.dot(inv_in.dot(np.transpose(g)))
num_out = g.dot(inv_in.dot(Cov2.dot(inv_in.dot(np.transpose(g)))))
for G in dominio:
  aux = G*G/(p/sigma)
  aux2 = G*G/deno_in[0][0]
  r_true.append(aux)
  r_in.append(aux2)
  r_out.append(G*G*num_out[0][0]/(deno_in[0][0]*deno_in[0][0]))
plt.plot(r_true,dominio,label="R^2_true")
plt.plot(r_in,dominio,label="R^2_in")
plt.plot(r_out,dominio,label="R^2_out")
plt.legend(loc='upper left',bbox_to_anchor = (1, .975)) 
```

```python

# Request historical data for past 5 years
dicc = ['JNJ','UNH','CHGCY','RHHVF','RHHBY',
        'PFE','TMO','NONOF','LLY','NVO',
        'DHR','ABT','ABBV','MRK','AZNCF',
        'TAK','AZN','HOCPY','NVS','MDT',
        'MRNA','BMY','LZAGY','SNY','AMGN',
        'ISRG','CVS','SYK','GLAXF','BAYRY',
        'GSK','CSLLY','MKKGY','DXCM','ZTS',
        'ANTM','GILD','IDXX','ESLOY','HCA',
        'SMMNY','BDX','EW','WXXWY','ALGN',
        'CI','BNTX','REGN','ILMN','BSX']
Precios = np.zeros(50*200)
Precios = Precios.reshape(50,200)
for i in range(len(dicc)):
  data = yf.Ticker(dicc[i]).history(period='200d')
  precios_stock = data.Close
  for j in range(200):
    Precios[i][j] = precios_stock[j]
Precios
```

```python
retornos  = np.zeros(50*199)
retornos = retornos.reshape(50,199)
for i in range(len(dicc)):
  for j in range(199):
    retornos[i][j] = (Precios[i][j+1]-Precios[i][j])/Precios[i][j]

retornos 
```

```python
dominio = np.linspace(0,100,1000)
r_in_clipping = []
r_out_clipping = []
eigenvalues, eigenvectors = np.linalg.eigh(Cov1)
eigenvalues2, eigenvectors2 = np.linalg.eigh(Cov2)
q = p/(n)
a1 = (1+np.sqrt(q))*(1+np.sqrt(q))
Covaux = np.zeros(p*p).reshape(p,p)
Covaux2 = np.zeros(p*p).reshape(p,p)
sum1 = 0
sum2 = 0
cont1 = 0
cont2 = 0
for i in range(p):
  if(eigenvalues[i]>=a1):
    sum1 = sum1 + eigenvalues[i]
    cont1 = cont1 + 1
  if(eigenvalues2[i]>=a1):
    sum2 = sum2 + eigenvalues2[i]
    cont2 = cont2 +1
gamma1 = 0
if(cont1<p):
  gamma1 = (p-sum1)/(p-cont1)
gamma2 = 0
if(cont2<p):
  gamma2 = (p-sum2)/(p-cont2)
for i in range(p):
  eigvector = eigenvectors[i].reshape(p,1)
  eigvector2 = eigenvectors2[i].reshape(p,1)
  if(eigenvalues[i]>=a1):
    Covaux = Covaux + eigenvalues[i]*eigvector.dot(np.transpose(eigvector))
  else :
    Covaux = Covaux + gamma1*eigvector.dot(np.transpose(eigvector))
  if(eigenvalues2[i]>=a1):
    Covaux2 = Covaux2 + eigenvalues2[i]*eigvector2.dot(np.transpose(eigvector2))
  else :
    Covaux2 = Covaux2 + gamma2*eigvector2.dot(np.transpose(eigvector2))
inv_in = np.linalg.inv(Covaux)
deno_in = g.dot(inv_in.dot(np.transpose(g)))
num_out = g.dot(inv_in.dot(Covaux2.dot(inv_in.dot(np.transpose(g)))))
for G in dominio:
  aux2 = G*G/deno_in[0][0]
  r_in_clipping.append(aux2)
  r_out_clipping.append(G*G*num_out[0][0]/(deno_in[0][0]*deno_in[0][0]))
```

```python
plt.title("Estimador clipping")
plt.plot(r_in,dominio,label="R^2_in")
plt.plot(r_out,dominio,label="R^2_out")
plt.plot(r_in_clipping,dominio,label="R^2_in,clipping")
plt.plot(r_out_clipping,dominio,label="R^2_out,clipping")
plt.legend(loc='upper left',bbox_to_anchor = (1, .975)) 
```
