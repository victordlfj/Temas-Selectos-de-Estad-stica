```python

#Definicion distribucion gaussiana analítica
def gaussian(x,mu,sigma):
    dist = np.zeros([len(x)])
    for i in range(len(x)):
        dist[i] = 1./np.sqrt(2*math.pi*sigma**2)*math.e**(-(x[i]-mu)**2/(2*sigma**2))
    return dist
points = 100000   
t = np.linspace(-5,5,points) 
plt.plot(t,gaussian(t,0,1),'-k',label='Atractor')

#Paramétro: suma de distribuciones
iteration = 2


# a) uniforme
a = 1
x = np.zeros([points])
for i in range(points):
    for k in range(iteration):
        x[i] += np.random.uniform(-a,a) 
        
sns.histplot(data=(x-np.mean(x))/np.std(x),stat="density", element="step", fill=False,kde=False,label='Uniforme')        


# b) exponencial
lam = 1
x = np.zeros([points])
for i in range(points):
    for k in range(iteration):
        x[i] += np.random.exponential(1./lam) 
sns.histplot(data=(x-np.mean(x))/np.std(x),stat="density", element="step", fill=False,kde=False,label='Exponencial')        


# c) normal
mu = 0
sigma = 1
x = np.zeros([points])
for i in range(points):
    for k in range(iteration):
        x[i] += np.random.normal(mu,sigma) 
sns.histplot(data=(x-np.mean(x))/np.std(x),stat="density", element="step", fill=False,kde=False,label='Gaussiana')   

plt.legend()
plt.show()
```

```python
#Calculamos la densidad de una mezcla de ensembles GOE y Wishart
from numpy import linalg as LA
#Función de la ley del semicírculo
def f(x):
    return 1/np.pi * np.sqrt(2-x**2) 

#Construimos la distribución de Marcenko-Pasteur
x = np.zeros(5000)
y = np.zeros(5000)
cont = 0
m = 100
p = 100
n = 200
c = p/n
delta_min = (1-np.sqrt(c)) ** 2
delta_max = (1+ np.sqrt(c)) ** 2
for i in np.linspace(delta_min,delta_max,100):
    x[cont] = i
    y[cont] = np.sqrt((delta_max-i)*(i-delta_min))/(2*np.pi*c*i)
    cont = cont +1
plt.plot(x,y,label='Marcenko Pasteur')
eps = 0.0001
t = np.arange(-np.sqrt(2)+eps, np.sqrt(2)-eps, 0.001)
plt.plot(t, f(t), 'k',linewidth=2,label=r'Semicirculo')

#Construimos la densidad del GOE+Wishart con r=.2 mediante simulación
eig_sum = []
k = 0
s = np.zeros([(n-1)*m])
r = .2
for i in range(m):
    #Creamos la matriz GOE
    H_GOE = np.sqrt(1/p)*np.random.normal(size=(p, p))
    Hsy_GOE = (H_GOE + np.transpose(H_GOE))/2.
    #Creamos la matriz Wishart
    H_Wishart = np.random.randn(p,n)
    Hsy_Wishart = (1/n)*H_Wishart.dot(np.transpose(H_Wishart))
    #Creamos una matriz del ensemble mezcla
    H_sum = r*Hsy_GOE + (1-r)*Hsy_Wishart
    w,v = LA.eigh(H_sum)  
    eig_sum.append(w)   
#Graficamos la densidad de eigenvalores del ensemble
GOE_Wishart = np.sort(np.array(eig_sum).flatten())
sns.histplot(data=GOE_Wishart,stat="density", element="step", fill=False, label='Simulacion')

#Calculamos la densidad numéricamente mediante un solver
from sympy.solvers import solve
from sympy.abc import g
p = 100
n = 200
#Inicializamos los parámetros w,q
w = .2
q = p/n
points = 100
densidad = []
#Para cada valor en el rango de la densidad calculamos el valor de f(z), la densidad
for z in np.linspace(-1.5,3,points):
  #despejamos el resolvente de la ecuación que lo relaciona con w,q,z
  sol = solve(w**2*g/2 + (1-w)/(1-q*(1-w)*g) + 1/g - z,g)
  #Encontramos el valor delr esolvente eligiendo la solución con parte imaginaria más grande
  maximag = 0
  for i in range(len(sol)):
    aux = complex(sol[i])
    if(aux.imag>maximag):
      maximag = aux.imag
  #Calculamos la densidad
  densidad.append(maximag/np.pi)
t = np.linspace(-1.5,3,points)
plt.plot(t,densidad,label="GOE-Wishart")
plt.title('r=.2')
plt.xlim([-1.5,4])
plt.ylim([0,1])
plt.legend(loc='upper left',bbox_to_anchor = (1, .975)) 
```
